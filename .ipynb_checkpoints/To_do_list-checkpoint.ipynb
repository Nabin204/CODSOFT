```python
# To do list 

# Importing modules

from tkinter import * 
from tkinter import messagebox

def add_task():
    task=my_entry.get()
    if (task!=""):
        lb.insert(END,task)
        my_entry.delete(0,"end")
    else:
        messagebox.showwarning("warning","Please enter a task")
def delete_task():
    lb.delete(ANCHOR)

root = Tk()
root.geometry("600x500+500+200")
root.title("To do list")
root.resizable(width=False,height=False)
root.config(bg="#223441")

title_frame=Frame(root)
title_frame.pack(side=TOP,pady=20)

frame=Frame(root)
frame.pack(pady=10)

label1=Label(title_frame,text="To-Do-List-App",fg="blue",border=4,font=('helvetica',30,'bold'))
label1.pack(fill=BOTH,side=TOP)

lb = Listbox(frame,width=25,height=8,font=("new times roman",18),bd=0,fg='#464646',highlightthickness=0,selectbackground="#a6a6a6",activestyle="none")
lb.pack(side=LEFT,fill=BOTH)

task_list=["reading book","running","watching movies","watering","bathing","playing cricket"]
for item in task_list:
    lb.insert(END,item)

sb=Scrollbar(frame)
sb.pack(side=RIGHT,fill=BOTH)
lb.config(yscrollcommand=sb.set)
sb.config(command=lb.yview)

my_entry=Entry(root,font=("times",24))
my_entry.pack(pady=20)
button_frame=Frame(root)
button_frame.pack(pady=20)

# Add button
add_task_btn=Button(button_frame,text="Add Task",font=("times",14),bg="#c5f776",padx=20,pady=10,command=add_task)
add_task_btn.pack(side=LEFT,fill=BOTH,expand=True)

# for deleting task simply select task and click on delete button

#Delete button

del_task_btn=Button(button_frame,text="Delete Task",font=("times",14),bg="#aa1010",padx=20,pady=10,command=delete_task)
del_task_btn.pack(side=RIGHT,fill=BOTH,expand=True)


root.mainloop()
```
