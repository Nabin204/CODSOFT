```python
#GUI Calculator Using Python 


from tkinter import * 
root=Tk()


def BtnClick(number):
    global var
    var=var+str(number)
    data.set(var)


def BtnClear():
    global var
    var=''
    data.set('')


def BtnEquals():
    global var
    result=str(eval(var))
    data.set(result)


root.title("Calculator")
root.geometry("361x381+500+100")
var=""
data=StringVar()
display=Entry(root,textvariable=data,font=("arial",20),bd=29,bg="powder blue",justify="right")
display.grid(row=0,columnspan=4)


btn7=Button(root,text="7",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(7))
btn7.grid(row=1,column=0)
btn8=Button(root,text="8",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(8))
btn8.grid(row=1,column=1)
btn9=Button(root,text="9",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(9))
btn9.grid(row=1,column=2)
btn_add=Button(root,text="+",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick('+'))
btn_add.grid(row=1,column=3)


btn4=Button(root,text="4",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(4))
btn4.grid(row=2,column=0)
btn5=Button(root,text="5",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(5))
btn5.grid(row=2,column=1)
btn6=Button(root,text="6",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(6))
btn6.grid(row=2,column=2)
btn_sub=Button(root,text="-",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick('-'))
btn_sub.grid(row=2,column=3)


btn1=Button(root,text="1",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(1))
btn1.grid(row=3,column=0)
btn2=Button(root,text="2",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(2))
btn2.grid(row=3,column=1)
btn3=Button(root,text="3",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(3))
btn3.grid(row=3,column=2)
btn_mul=Button(root,text="*",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick('*'))
btn_mul.grid(row=3,column=3)


btnC=Button(root,text="C",font=("arial",12,"bold"),height=2,width=6,bd=12,command=BtnClear)
btnC.grid(row=4,column=0)
btn0=Button(root,text="0",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick(0))
btn0.grid(row=4,column=1)
btn_div=Button(root,text="/",font=("arial",12,"bold"),height=2,width=6,bd=12,command=lambda:BtnClick('/'))
btn_div.grid(row=4,column=2)
btn_equal=Button(root,text="=",font=("arial",12,"bold"),height=2,width=6,bd=12,command=BtnEquals)
btn_equal.grid(row=4,column=3)


root.mainloop()
```
